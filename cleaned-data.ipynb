```python
import pandas as pd
csv_file = "city_day.csv"
file_path = os.path.join(path, csv_file)
df = pd.read_csv(file_path)
print(df.head())
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.isna().sum()
```

```python
df.loc[:, 'PM2.5'] = df['PM2.5'].fillna(df['PM2.5'].mean().round(2))
df.loc[:, 'PM10'] = df['PM10'].fillna(df['PM10'].mean().round(2))
df.loc[:, 'NO'] = df['NO'].fillna(df['NO'].mean().round(2))
df.loc[:, 'NO2'] = df['NO2'].fillna(df['NO2'].mean().round(2))
df.loc[:, 'NOx'] = df['NOx'].fillna(df['NOx'].mean().round(2))
df.loc[:, 'NH3'] = df['NH3'].fillna(df['NH3'].mean().round(2))
df.loc[:, 'CO'] = df['CO'].fillna(df['CO'].mean().round(2))
df.loc[:, 'SO2'] = df['SO2'].fillna(df['SO2'].mean().round(2))
df.loc[:, 'O3'] = df['O3'].fillna(df['O3'].mean().round(2))
df.loc[:, 'Benzene'] = df['Benzene'].fillna(df['Benzene'].mean().round(2))
df.loc[:, 'Toluene'] = df['Toluene'].fillna(df['Toluene'].mean().round(2))
df.loc[:, 'Xylene'] = df['Xylene'].fillna(df['Xylene'].mean().round(2))
```

```python
df
```

```python
df['AQI'].fillna(df['AQI'].median(), inplace=True)
```

```python
df['AQI_Bucket'].isna().sum()
```

```python
from sklearn.impute import KNNImputer
# Selecting relevant columns for KNN Imputation
columns_to_impute = ['PM2.5', 'PM10', 'NO2', 'SO2', 'O3', 'AQI']

# Applying KNN Imputer
imputer = KNNImputer(n_neighbors=5)
df[columns_to_impute] = imputer.fit_transform(df[columns_to_impute])
```

```python
def assign_aqi_bucket(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    else:
        return 'Severe'

df['AQI_Bucket'] = df['AQI'].apply(assign_aqi_bucket)
```
